```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data1 = np.random.normal(0, 1, 1000)  # Normal distribution
data2 = np.random.normal(2, 0.5, 1000)  # Normal distribution with different mean and std

# Create a figure and axis
fig, ax = plt.subplots()

# Create a histogram for the first dataset
ax.hist(data1, bins=30, alpha=0.5, color='blue', label='Data 1', density=True, histtype='step')

# Create a histogram for the second dataset
ax.hist(data2, bins=30, alpha=0.5, color='red', label='Data 2', density=True, histtype='stepfilled')

# Add labels and title
ax.set_xlabel('Value')
ax.set_ylabel('Density')
ax.set_title('Histogram Example with ax.hist()')
ax.legend()

# Show the plot
plt.show()
```
