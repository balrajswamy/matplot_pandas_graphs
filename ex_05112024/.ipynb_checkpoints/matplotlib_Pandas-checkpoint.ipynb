```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df = pd.read_csv("cars_data.csv")

date_col = pd.date_range(start="2024-10-12", periods=5)
#df = pd.DataFrame(date_col, columns=["Date"])
df["Date"] = date_col
df.columns
```

```python
df
```

```python
df["Price"]=df["Price"].astype(float)
```

```python
df["Date"] = df["Date"].astype(str)
```

```python
df.plot(x = "Date", y = "Price",figsize=(10, 5), kind="bar")
plt.title('Sales Over Years')
plt.ylabel('Sales')
plt.xlabel('Year')

plt.show()
```
