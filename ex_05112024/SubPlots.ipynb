```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

x = [1,2,3,4,5]
y = [6,7,8,9,10]

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2,figsize=(10,10))
ax1.plot(x,y)
ax2.plot(x,y)
ax1.set(title="Plots shown as twice", xlabel="Numbers", ylabel="Items")
ax2.set(title="Plots shown as twice", xlabel="Numbers", ylabel="Items")

plt.show()

           
     
```
