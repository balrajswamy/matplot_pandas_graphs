```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

marks = [15,33,40,32,65,68,72,81,76,79,26,59,81,87,99,49,76]
grade_intervals =[0,35,65,80,100]
fig, ax = plt.subplots()
ax.set(title="Marks of the Students", xlabel="Marks", ylabel = "No_of_Students",xticks= grade_intervals)
#ax.hist(marks, grade_intervals, fc="green", ec="red",label="marks")
ax.hist(x = marks, bins = grade_intervals, fc="green", ec="red",label="marks")
ax.legend(["students","marks"], loc="upper left")
plt.show()
```
